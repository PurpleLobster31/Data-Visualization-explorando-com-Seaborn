# Tradução 

```python
dados.columns
```

```python
colunas_renomeadas = {
    'total_bill': 'valor_da_conta',
    'tip': 'gorjeta',
    'dessert': 'sobremesa',
    'day': 'dia_da_semana',
    'time': 'hora_do_dia',
    'size': 'total_de_pessoas'
}
```

```python
gorjetas = dados.rename(columns=colunas_renomeadas)
```

```python
gorjetas
```

```python
gorjetas.sobremesa.unique()
```

```python
sim_nao = {
    'Yes': 'Sim',
    'No': 'Nao'
}
```

```python
gorjetas.sobremesa.map(sim_nao)
```

```python
gorjetas.head()
```

```python
gorjetas.sobremesa = gorjetas.sobremesa.map(sim_nao)
```

```python
gorjetas.head()
```

```python
gorjetas.dia_da_semana.unique()
```

```python
dias = {
    'Sun': 'Dom',
    'Sat': 'Sab',
    'Thur': 'Qui',
    'Fri': 'Sex'
}
```

```python
gorjetas.dia_da_semana = gorjetas.dia_da_semana.map(dias)
```

```python
gorjetas.head()
```

```python
gorjetas.hora_do_dia.unique()
```

```python
hora = {
    'Dinner': 'Jantar',
    'Lunch': 'Almoço'
}
```

```python
gorjetas.hora_do_dia = gorjetas.hora_do_dia.map(hora)
```
