## Teste de Hipótese

**H<sup>null</sup>**
> **A distribuição da taxa da gorjeta é a mesma nos dois grupos**

**H<sup>alt</sup>**
> **A distribuição da taxa da gorjeta não é a mesma nos dois grupos**

```python
from scipy.stats import ranksums
```

```python
sobremesa = gorjetas.query("sobremesa == 'Sim'")['porcentagem']
```

```python
sem_sobremesa = gorjetas.query("sobremesa == 'Nao'")['porcentagem']
```

```python
r = ranksums(sobremesa, sem_sobremesa)
```

```python
print('O valor do pvalue é: {}'.format(r.pvalue))
```

> **A distribuição da taxa da gorjeta é a mesma nos dois grupos**

## Teste de Hipótese

**H<sup>null</sup>**
> **A distribuição do valor da conta é igual no sábado e no domingo**

**H<sup>alt</sup>**
> **A distribuição do valor da conta não é igual no sábado e no domingo**

```python
valor_conta_sabado = gorjetas.query("dia_da_semana == 'Sab'")['valor_da_conta']
```

```python
valor_conta_domingo = gorjetas.query("dia_da_semana == 'Dom'")['valor_da_conta']
```

```python
r2 = ranksums(valor_conta_sabado, valor_conta_domingo)
```

**H<sup>null</sup>**
> **A distribuição do valor da conta é igual no sábado e no domingo**

```python
print('O valor do pvalue é: {}'.format(r2.pvalue))
```
